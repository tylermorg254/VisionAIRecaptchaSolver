```python
from ultralytics import YOLO

def train():
    
    model = YOLO("yolo11x-cls.pt")

    results = model.train(
        data="dataset_cls_full_57k",
        epochs=50,
        imgsz=640,
        batch=64,
        device=0,
        workers=64,
        patience=15,
        project="runs/classify",
        name="rec_cls_model",
        amp=True,
        cache=True,
    )

if __name__ == "__main__":
    train()
```

```python
from ultralytics import YOLO

def resume_training():

    path_to_last_weights = "runs/classify/rec_cls_model/weights/best.pt"

    try:
        model = YOLO(path_to_last_weights)

        results = model.train(resume=True)

    except FileNotFoundError:
        print(f"Error: file not found {path_to_last_weights}")

if __name__ == "__main__":
    resume_training()
```
